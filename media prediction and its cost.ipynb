```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

print("Libraries imported successfully!")
```

```python
df = pd.read_csv('media prediction and its cost.csv')

# Check first 5 rows
df.head()
```

```python
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nMissing Values:")
print(df.isnull().sum())
print("\nBasic Stats:")
df.describe()
```

```python
# Drop rows with missing values
df = df.dropna()

# Reset index
df = df.reset_index(drop=True)

print("Data cleaned!")
print("Total Records:", len(df))
```

```python
# Define funnel stages using store sales data
funnel_data = {
    'Stage': ['Awareness\n(Total Customers)', 
              'Interest\n(Made a Purchase)', 
              'Consideration\n(Repeat Purchase)', 
              'Conversion\n(High Value Purchase)'],
    'Count': [
        len(df),
        len(df[df['store_sales(in millions)'] > 0]),
        len(df[df['store_sales(in millions)'] > df['store_sales(in millions)'].median()]),
        len(df[df['store_sales(in millions)'] > df['store_sales(in millions)'].quantile(0.75)])
    ]
}

funnel_df = pd.DataFrame(funnel_data)
print(funnel_df)
```

```python
plt.figure(figsize=(10,6))
colors = ['#2ecc71','#3498db','#e67e22','#e74c3c']

bars = plt.barh(funnel_df['Stage'], funnel_df['Count'], color=colors)

# Add value labels
for bar, val in zip(bars, funnel_df['Count']):
    plt.text(bar.get_width() + 50, bar.get_y() + bar.get_height()/2,
             f'{val:,}', va='center', fontweight='bold')

plt.title('Marketing Funnel - Customer Journey', fontsize=14)
plt.xlabel('Number of Customers')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig('marketing_funnel.png')
plt.show()
```

```python
total = funnel_df['Count'][0]

funnel_df['Conversion Rate %'] = (funnel_df['Count'] / total * 100).round(2)
funnel_df['Drop Off %'] = (100 - funnel_df['Conversion Rate %']).round(2)

print("===== FUNNEL CONVERSION RATES =====")
print(funnel_df[['Stage','Count','Conversion Rate %']].to_string(index=False))
```

```python
drop_offs = []
for i in range(1, len(funnel_df)):
    prev = funnel_df['Count'][i-1]
    curr = funnel_df['Count'][i]
    drop = ((prev - curr) / prev * 100)
    drop_offs.append(round(drop, 1))

drop_offs.insert(0, 0)  # No drop off at first stage
funnel_df['Drop Off Between Stages %'] = drop_offs

plt.figure(figsize=(10,5))
plt.bar(funnel_df['Stage'][1:], drop_offs[1:], color='tomato')
plt.title('Drop-Off Rate Between Funnel Stages (%)')
plt.xlabel('Stage Transition')
plt.ylabel('Drop Off (%)')
plt.tight_layout()
plt.savefig('dropoff_analysis.png')
plt.show()

print("\nBiggest Drop Off Stage:")
max_drop_idx = drop_offs.index(max(drop_offs[1:]))
print(funnel_df['Stage'][max_drop_idx])
```

```python
media_cols = ['radio ads', 'bulk mail', 'newspaper', 'TV ads']

# Check which media columns exist in dataset
available_media = [col for col in media_cols if col in df.columns]
print("Available media columns:", available_media)

if available_media:
    media_sales = df[available_media].sum().sort_values(ascending=False)
    
    plt.figure(figsize=(9,5))
    media_sales.plot(kind='bar', color=['steelblue','coral','green','orange'])
    plt.title('Marketing Channel Performance (Total Spend)')
    plt.xlabel('Channel')
    plt.ylabel('Total Amount')
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.savefig('channel_performance.png')
    plt.show()
    
    print("\nBest Performing Channel:", media_sales.idxmax())
```

```python
if 'store_sales(in millions)' in df.columns and 'store_cost(in millions)' in df.columns:
    df['Profit'] = df['store_sales(in millions)'] - df['store_cost(in millions)']
    df['Profit Margin %'] = (df['Profit'] / df['store_sales(in millions)'] * 100).round(2)

    plt.figure(figsize=(8,5))
    sns.histplot(df['Profit Margin %'], bins=30, color='steelblue', kde=True)
    plt.title('Distribution of Profit Margins Across Stores')
    plt.xlabel('Profit Margin (%)')
    plt.ylabel('Number of Stores')
    plt.tight_layout()
    plt.savefig('profit_margin.png')
    plt.show()

    print("Average Profit Margin:", df['Profit Margin %'].mean().round(2), "%")
```

```python
food_cols = [col for col in df.columns if 'food' in col.lower()]

if food_cols:
    # Convert all food columns to numbers, text becomes NaN
    df[food_cols] = df[food_cols].apply(pd.to_numeric, errors='coerce')
    
    # Fill NaN with 0
    df[food_cols] = df[food_cols].fillna(0)
    
    # Now sum and plot
    food_sales = df[food_cols].sum().sort_values(ascending=False).head(10)
    
    plt.figure(figsize=(12,5))
    food_sales.plot(kind='bar', color='mediumseagreen')
    plt.title('Top 10 Food Categories by Sales')
    plt.xlabel('Category')
    plt.ylabel('Total Sales')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig('food_categories.png')
    plt.show()
    
    print("Top Food Category:", food_sales.idxmax())
else:
    print("No food columns found in dataset!")
```

```python
if 'unit_sales(in millions)' in df.columns:
    df['Sales_Category'] = pd.cut(df['unit_sales(in millions)'],
                                   bins=4,
                                   labels=['Low','Medium','High','Very High'])

    sales_cat = df['Sales_Category'].value_counts()

    plt.figure(figsize=(7,7))
    plt.pie(sales_cat, labels=sales_cat.index,
            autopct='%1.1f%%',
            colors=['tomato','orange','steelblue','green'],
            startangle=90)
    plt.title('Customer Distribution by Sales Volume')
    plt.tight_layout()
    plt.savefig('sales_distribution.png')
    plt.show()
```

```python
print("========== KEY FINDINGS ==========\n")

print(f"1. Total Records Analyzed: {len(df):,}")

conversion_rate = (funnel_df['Count'].iloc[-1] / funnel_df['Count'].iloc[0] * 100)
print(f"2. Overall Funnel Conversion Rate: {conversion_rate:.1f}%")

biggest_drop = funnel_df.loc[funnel_df['Drop Off Between Stages %'].idxmax(), 'Stage']
print(f"3. Biggest Drop-Off Stage: {biggest_drop}")

if available_media:
    print(f"4. Best Marketing Channel: {media_sales.idxmax()}")

if 'Profit Margin %' in df.columns:
    print(f"5. Average Profit Margin: {df['Profit Margin %'].mean():.1f}%")
```
